```python
import numpy as np

A = np.array([[0.3, 0.2, 0.5],
              [0.6, 0.3, 0.1],
              [0.2, 0.4, 0.4]])

B = np.array([[0.7, 0.1, 0.2],
              [0.5, 0.4, 0.1],
              [0.6, 0.1, 0.3]])

C = np.array([[0.2, 0.3, 0.3],
              [0.1, 0.5, 0.4],
              [0.4, 0.4, 0.2]])


DP = np.array([[0.3, 0.5, 0.2],
              [0.6, 0.2, 0.2],
              [0.5, 0.1, 0.4]])

# Calculating the squared Euclidean distance
dist_squared_1 = np.sum((A - DP) ** 2)
dist_squared_2 = np.sum((B - DP) ** 2)
dist_squared_3 = np.sum((C - DP) ** 2)
min = np.min(np.array([dist_squared_1, dist_squared_2, dist_squared_3]))
print(f"Squared Euclidean distance: {min}")
```
